```python
dic = []
with open('dict.txt', 'r', encoding='utf-8') as file:
    for line in file:
        word = line.strip().split(' ')[0]  # 使用空格分割并只取第一个部分
        dic.append(word)

print(dic[:5])
```

```python
#實作完全切分的function
def fully_segment(text,dic):
  word_list = []
  for i in range(len(text)):
    for j in range(i+1,len(text)+1):
      word = text[i:j]
      if word in dic:
        word_list.append(word)
  return word_list
```

```python
#實作正向最長匹配function
def forward_segment(text, dic):
    word_list = []
    i = 0
    while i < len(text):
        longest_word = text[i]  # 注意這裡從i開始
        for j in range(i + 1, len(text) + 1):
            word = text[i:j]
            if word in dic:
                if len(word) > len(longest_word):
                    longest_word = word
        word_list.append(longest_word)
        i += len(longest_word)
    return word_list
```

```python
#實作逆向最長匹配function
def backward_segment(text, dic):
    word_list = []
    i = len(text) - 1
    while i >= 0:
        longest_word = text[i]
        for j in range(i, -1, -1):
            word = text[j:i + 1]
            if word in dic:
                if len(word) > len(longest_word):
                    longest_word = word
        word_list.insert(0, longest_word)
        i -= len(longest_word)
    return word_list

```

```python
#實作雙向最長匹配function
def count_single_char(word_list: list):
    return sum(1 for word in word_list if len(word) == 1)

def bidirectional_segment(text, dic):
    f = forward_segment(text, dic)
    b = backward_segment(text, dic)

    if len(f) < len(b):
        return f
    elif len(f) > len(b):
        return b
    else:
        if count_single_char(f) < count_single_char(b):
            return f
        else:
            return b
```

```python
#使用完全切分法來切十句話，並儲存結果為一個double List以便後續處理

fullySegmentResults = []
for i in range(len(cleanTextLib)):
  segmentResult = fully_segment(cleanTextLib[i],dic)
  fullySegmentResults.append(segmentResult)
print(fullySegmentResults)
```

```python
#使用正向最長匹配切分法來切十句話，並儲存結果為一個double List以便後續處理
forwardSegmentResults = []
for i in range(len(cleanTextLib)):
  segmentResult = forward_segment(cleanTextLib[i],dic)
  forwardSegmentResults.append(segmentResult)
print(forwardSegmentResults)
```

```python
#使用逆向最長匹配切分法來切十句話，並儲存結果為一個double List以便後續處理
backwardSegmentResults = []
for i in range(len(cleanTextLib)):
  segmentResult = backward_segment(cleanTextLib[i],dic)
  backwardSegmentResults.append(segmentResult)
print(backwardSegmentResults)
```

```python
#使用雙向最長匹配切分法來切十句話，並儲存結果為一個double List以便後續處理
bidirectionalSegmentResults = []
for i in range(len(cleanTextLib)):
  segmentResult = bidirectional_segment(cleanTextLib[i],dic)
  bidirectionalSegmentResults.append(segmentResult)
print(bidirectionalSegmentResults)
```

```python
def to_region(segmentation: str) -> list:
     """
     將分詞結果轉換爲區間
     """
     region = []
     start = 0
     for word in re.compile("\\s+").split(segmentation.strip()):
         end = start + len(word)
         region.append((start, end))
         start = end
     return region


def prf(gold: str, pred: str, dic) -> tuple:
     """
     計算P、R、F1
     :param gold: 標準答案文件
     :param pred: 分詞結果文件
     :param dic: 詞典
     :return: (P, R, F1, OOV_R, IV_R)
     """
     A_size, B_size, A_cap_B_size, OOV, IV, OOV_R, IV_R = 0, 0, 0, 0, 0, 0, 0
     A, B = set(to_region(gold)), set(to_region(pred))
     A_size += len(A)
     B_size += len(B)
     A_cap_B_size += len(A & B)
     text = re.sub("\\s+", "", gold)


     for (start, end) in A:
         word = text[start: end]
         if word in dic:
             IV += 1
         else:
             OOV += 1

     for (start, end) in A & B:
         word = text[start: end]
         if word in dic:
             IV_R += 1
         else:
             OOV_R += 1
     p, r = A_cap_B_size / B_size * 100, A_cap_B_size / A_size * 100
     f1 = 2 * p * r / (p + r) if (p + r) != 0 else 0
     oov_rate = OOV_R / OOV * 100 if OOV != 0 else 0
     iv_rate = IV_R / IV * 100 if IV != 0 else 0
     return p, r, 2 * p * r / (p + r), oov_rate, iv_rate
```

```python
#比較標準答案和整個segment方法的準確率（取平均)
def prf2(goldList,segmentResult) -> tuple:
  result = (0,0,0,0,0)
  for i, sublist in enumerate(segmentResult):
    segmentResult_str =" ".join(sublist)
    #print(segmentResult_str)
    gold = " ".join(goldList[i])
    #print(gold)
    t = prf(gold,segmentResult_str,dic)
    #print(t)



    result = [x + y for x, y in zip(result, t)]
  result = tuple(item / 10 for item in result)
  return result




```
