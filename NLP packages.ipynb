```python
# 引入所需的模組
import jieba  # jieba是一個中文分詞工具
import time   # time模組提供各種與時間有關的功能
import pandas as pd  # pandas是一個提供數據分析功能的Python庫
import re
```

```python

# 使用pandas讀取CSV
df = pd.read_csv("comment_ge20.csv")
name = []

#將其值加到name中
for i in df['MESSAGE']:
    name.append(i)


# 從name列表中取前1萬條數據作為樣本數據
sampled_data = name[:10000]

# 打印樣本數據的前5個值，確認樣本選取正確
print(sampled_data[:5])
```

```python
# 紀錄開始時間
start_time = time.time()

# 使用jieba進行斷詞。使用str()函數確保sentence是字符串
# 使用'/'作為分隔符將斷詞的結果連接起來
jieba_results = ['/'.join(jieba.cut(str(sentence))) for sentence in sampled_data]

# 計算使用jieba進行斷詞的所需時間
jieba_time = time.time() - start_time

# 打印結果
print(jieba_results)
```

```python
# 使用pip安裝thulac套件
!pip install thulac

# 導入thulac模組
import thulac

# 創建一個thulac的實例。使用seg_only=True表示只進行分詞，不進行詞性標註。
thu1 = thulac.thulac(seg_only=True)
```

```python
# 記錄當前時間，以計算分詞所需時間
start_time = time.time()

# 使用thulac進行分詞，並將分詞結果保存在segmented_data列表中。
# 對於sampled_data中的每一條句子，我們首先將其轉換為字符串格式，然後使用thulac進行分詞。
# text=True表示將分詞結果作為一個字符串返回。
segmented_data = [thu1.cut(str(sentence), text=True) for sentence in sampled_data]

# 計算所需的時間
tulac_time = time.time() - start_time

```

```python
# 使用pip工具安裝snownlp庫
!pip install snownlp
# 從snownlp庫中引入SnowNLP，這是一個用於中文自然語言處理的庫
from snownlp import SnowNLP
```

```python
# 記錄當前時間，用於計算snownlp的運行時間
start_time = time.time()

# 使用SnowNLP進行中文斷詞。這裡使用列表解析式對sampled_data中的每個句子進行斷詞，
# 斷詞的結果是一個詞語列表，然後將這些詞語使用'/'連接成一個字符串。
snownlp_result = ['/'.join(SnowNLP(str(sentence)).words) for sentence in sampled_data]

# 計算斷詞過程所花費的時間
snownlp_time = time.time() - start_time

# 打印斷詞後的結果中的前五個，以便觀察
print(snownlp_result[:5])


```

```python
def to_region(segmentation: str) -> list:
     """
     將分詞結果轉換爲區間
     """
     region = []
     start = 0
     for word in re.compile("\\s+").split(segmentation.strip()):
         end = start + len(word)
         region.append((start, end))
         start = end
     return region


def prf(gold: str, pred: str) -> tuple:
     """
     計算P、R、F1
     :param gold: 標準答案文件
     :param pred: 分詞結果文件
     :return: (P, R, F1)
     """
     A_size, B_size, A_cap_B_size, OOV, IV, OOV_R, IV_R = 0, 0, 0, 0, 0, 0, 0
     A, B = set(to_region(gold)), set(to_region(pred))
     A_size += len(A)
     B_size += len(B)
     A_cap_B_size += len(A & B)
     text = re.sub("\\s+", "", gold)
     p, r = A_cap_B_size / B_size * 100, A_cap_B_size / A_size * 100
     f1 = 2 * p * r / (p + r) if (p + r) != 0 else 0
     return p, r, f1
```

```python
#比較標準答案和整個segment方法的準確率（取平均)
def prf2(goldList,segmentResult) -> tuple:
  result = (0,0,0)
  for i, sentences in enumerate(segmentResult):
    words = re.findall(r'[^/]+', sentences)
    Sresult = ' '.join(words)
    segmentResult_str = Sresult
    print(segmentResult_str)
    gold = " ".join(goldList[i])
    print(gold)
    t = prf(gold,segmentResult_str)
    print(t)
    result = [x + y for x, y in zip(result, t)]
  result = tuple(item / 4 for item in result)
  return result
```
